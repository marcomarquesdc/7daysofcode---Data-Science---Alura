## Importação do arquivo
De acordo com o arquivo *README* da pasta *ml-100k* mostra que o *data set* que contém as informações principais é *u.data*, que mostra as recomentações dos usuários para os filmes então vou escolher um usuário para recomendar filmes com base nas suas avaliações.

### Visualizando dados
**Agora irei ler os dados e nomear as colunas pois o *data set* veio sem colunas, mas é possível nomeá-las, pois o *README* diz o nome das colunas de cada *data set*.**

```python
import pandas as pd

colunas = ["user id", "item id", "rating", "timestamp"]
dados = pd.read_table("../Day4/ml-100k/u.data", names=colunas)
dados
```

## Agrupando usuários
**Para fazer um sistema de recomendação irie analisar a avaliação de um usuário qualquer e analisar os tipos de filmes que esse usuário gosta, logo em seguida recomendar filmes que ele não tenha avaliado ainda baseado nos gêneros que ele goste.**

```python
selecao_usuario_196 = dados["user id"] == 196
df_user196 = dados[selecao_usuario_196]
df_user196
```

## Filtrando dados

**Agora vou filtrar para selecionar os filmes com avaliação acima de 3.**

```python
selecao_maior3 = df_user196["rating"] >= 3
df_user196_maior3 = df_user196[selecao_maior3]
df_user196_maior3
```

**Como não temos os nomes dos filmes teremos que ler outro *data set* com as especificações de cada filme.**

```python
colunas_filmes =  ["movie id", "movie title", "release date", "video release date",
              "IMDb URL", "unknown", "Action", "Adventure", "Animation",
              "Children's", "Comedy", "Crime", "Documentary", "Drama", "Fantasy",
              "Film-Noir", "Horror", "Musical", "Mystery", "Romance", "Sci-Fi",
              "Thriller", "War", "Western"]
filmes = pd.read_table("ml-100k/u.item", encoding="latin-1", sep="|", names=colunas_filmes)
filmes
```

**Na célula abaixo criarei uma seleção para filtrar, apenas para mostrar os filmes que o nosso usuário avaliou maior que 3.**

```python
selecao_filmes_user196 = filmes["movie id"].isin(df_user196_maior3["item id"])
filmes_do_user196 = filmes[selecao_filmes_user196]
filmes_do_user196
```

**Com essa filtragem podemos ver os nomes dos filmes que o *user196* avaliou maior que 3**

## Analisando filmes do user 196

**A célula abaixo contém a soma dos gêneros dos filmes que o *user 196* avaliou maior que 3.**

```python
filmes_do_user196.loc[:,"unknown":].sum()
```

**Como podemos ver o *user 196* gosta majoritariamente de filmes de comédia e drama, logo irei recomendar filmes de comédia e/ou de drama.**

## Recomendando filmes

**Aqui vou fazer um seleção de filmes que o *user_196* não viu e depois filtrá-la para filmes de comédia e romance.**

```python
selecao_filmes_naovistos = filmes["movie id"].isin(df_user196["item id"].unique())
df_filmes_naovistos = filmes[~selecao_filmes_naovistos]
df_filmes_naovistos
```

**Agora vou filtrar para mostrar apenas filmes de comédia ou romance.**

```python
selecao_romanceoucomedia = (df_filmes_naovistos["Romance"] == 1) | (df_filmes_naovistos["Comedy"] == 1)
df_filmes_naovistos[selecao_romanceoucomedia]
```

**Agora vou filtrar para mostrar apenas filmes de comédia e romance.**

```python
selecao_romanceecomedia = (df_filmes_naovistos["Romance"] == 1) & (df_filmes_naovistos["Comedy"] == 1)
df_filmes_naovistos[selecao_romanceecomedia]
```
