## Analisando e tratando colunas 

**Na célula de código abaixo ireiverificar o tipo de variável de cada coluna com o método *d_types*.**

```python
dados.dtypes
```

**Podemos ver que a maioria das colunas tem valores do tipo *object*, ou seja, muitas entradas tem tipos de dados misturados na mesma coluna, para resolver isso a melhor opção é mudar o tipo de variável para uma só.**<br><br>**Na célula de código abaixo identifiquei na coluna *VALOR_REEMBOLSADO* todas as entradas que eram do tipo *stirng* e onde tinha  ','  foi trocado por um  '.'  para fazer a troca de tipo de variável, que no caso foi trocada para *float*.**

```python
lista = []
strings = []
floats = []

#Identificando e trocando ',' por '.'
for v in dados["VALOR_REEMBOLSADO"] :
    if type(v) == str :
        v = list(v)
        for l in range(len(v)) :
            if v[l] == "," :
                v[l] = "."
                lista.append(v)
                
#Concatenando as strings das listas
for x in lista :
    s = "".join(x)
    strings.append(s)

#Transformando em float
for n in strings :
    n = float(n)
    floats.append(n)
```

**Na célula de código abaixo irei identificar na coluna *VALOR_REEMBOLSADO* qual valores são do tipo *string* e trocar pelas novas variáveis do tipo *float*, assim substituindo tudo por valores numéricos**

```python
valores = []
novos_valores = []
valores_float = []
ifl = 0

for valor in dados["VALOR_REEMBOLSADO"] :
    valores.append(valor)
for iv in range(len(valores)) :
    if type(valores[iv]) != str :
        novos_valores.append(valores[iv])
    else:
        valores[iv] = floats[ifl]
        novos_valores.append(valores[iv])
        ifl += 1
for q in novos_valores :
    q = float(q)
    valores_float.append(q)

valores_float = np.array(valores_float)
```

**Agora irei criar uma nova coluna chamada de *VALOR* com apenas entradas numéricas, assim poderei fazer operações matemáticas com ela, também remover a coluna *VALOR_REEMBOLSADO* que é inútil para fazer análises e utilizar o método *drop_duplicates* para ver se há dados repetidos.**

```python
dados["VALOR"] = valores_float
dados = dados.drop(["VALOR_REEMBOLSADO"], axis=1)
dados = dados.drop_duplicates()
dados
```
